# Thresholding dense data: have a dictionary to record number of L domains recorded used (so far), if exceeds 50, then we skip it for the resulting list.

```python
df_shuff = df_nodes.sample(frac=1).reset_index(drop=True)
df_shuff
```

```python
regulating_dict = {}
skimed_column = []
for i, row in df_shuff.iterrows():
  skimed_pl = []
  cur_pl = row.get('PplusL')
  for ele in cur_pl:
    if ele not in regulating_dict:
      skimed_pl.append(ele)
      regulating_dict[ele]=1
    else:
      if regulating_dict[ele] <= 50:
        skimed_pl.append(ele)
        regulating_dict[ele]+=1
      else:
        continue
  skimed_column.append(skimed_pl)
df_shuff['skimed_pl'] = skimed_column
df_shuff  
```

## Sovling Sparse Problem: N-minsupport recursive
We want to make sure each meme has at least 5 connected domains, and each domain has at least 5 connected memes. 

```python
def filter_memes(df, min_support):
  df['support'] = df.apply(lambda x: len(x.skimed_pl), axis=1)
  df = df.loc[df['support'] >= min_support]
  df = df[['Q', 'skimed_pl']].reset_index()
  return df

# df1 = filter_memes(df_shuff, 4)
# df1
```

```python
def filter_domains(df, min_support):
  df = df.explode('skimed_pl')
  dftemp = df.groupby(['skimed_pl']).size().reset_index(name='counts')
  dftemp = dftemp.loc[dftemp['counts']>=min_support]
  kept_pls = list(dftemp['skimed_pl'])
  print("num of kept domains for min support= {} is {}".format(min_support,len(kept_pls)))
  df = df.loc[df['skimed_pl'].isin(kept_pls)]
  df_res = (df.groupby(['Q']).agg({'skimed_pl': lambda x: x.tolist()}).reset_index())
  return df_res
```

# change the min_support number to alter the min_support filtering

```python
prev_shape = df_shuff.shape
df2_shape = None
df2 = df_shuff
while df2_shape != prev_shape:
  print(df2_shape)
  prev_shape = df2_shape
  df1 = filter_memes(df2, 3)
  df1
  df2 = filter_domains(df1, 3)
  df2_shape = df2.shape
```
